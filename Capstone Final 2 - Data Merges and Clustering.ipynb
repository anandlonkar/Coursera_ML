# Merge data

Combine all data based on Zip codes

```python
MergedDF = GroupedDF.merge(Zip_Age_data_filtered)
MergedDF = MergedDF.merge(Zip_housing_data_filtered)
MergedDF = MergedDF.merge(Zip_income_data_filtered)
MergedDF.head()
```

Calculate kids to preschool ratio. This is used later to find the best zip code

```python
MergedDF['kids_to_PreSchool']=MergedDF['Population_Under_5']/MergedDF['Count_of_preschools']
MergedDF.head()
```

Get data in numeric form to support Kmeans

```python
MergedDF.dtypes
MergedDF = MergedDF.apply(pd.to_numeric,errors='coerce')
MergedDF.dtypes
MergedDF.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
MergedDF.to_csv('MergedDF.csv')
```

Fit data for Kmeans clustering

```python
from sklearn.cluster import KMeans
# set number of clusters
kclusters = 5

#MergedDF = MergedDF.drop('Zipcode', 1)

# run k-means clustering
kmeans = KMeans(n_clusters = kclusters, init = 'k-means++', max_iter=1000, n_init = 100, random_state=0).fit_predict(MergedDF)

# check cluster labels generated for each row in the dataframe
kmeans
```

```python
MergedDF.insert(0, 'Cluster Labels', kmeans)
```

Sort by the kids to school ratio to find best zip code

```python
BestCluster = MergedDF[MergedDF['Cluster Labels']==2].sort_values('kids_to_PreSchool',ascending=False).head(n=1)
```

```python
BestCluster
```

```python
BestZip_Lat = BestCluster['Latitude']
BestZip_Lng = BestCluster['Longitude']
```

# Display on map

```python
# create map
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors
map_clusters = folium.Map(location=[33.155373, -96.818733], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(MergedDF['Latitude'], MergedDF['Longitude'], MergedDF['Zipcode'], MergedDF['Cluster Labels']):
    
    if (lat != float(BestZip_Lat) and lon != float(BestZip_Lng)):
        label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color=rainbow[cluster-1],
            fill=True,
            fill_color=rainbow[cluster-1],
            fill_opacity=0.7).add_to(map_clusters)
    else:
        label = folium.Popup(str(poi) + ' BEST LOCATION! Cluster ' + str(cluster), parse_html=True)
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color='red',
            fill=True,
            fill_color=rainbow[cluster-1],
            fill_opacity=0.7).add_to(map_clusters)

map_clusters
```
